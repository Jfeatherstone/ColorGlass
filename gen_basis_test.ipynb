# Generalized Basis Test

In this file, we verify that the generalized matrices computed for the arbitrary color wavefunctions satisfy the requisite property:

$$ tr( t^a t^b) = \frac{1}{2} \delta_{ab} $$

```python
import numpy as np
import matplotlib.pyplot as plt

import cgc
```

```python
# We ignore the last element since that will be the identity
pauliMatrices = cgc.ArbColors.get_basis(2)[:-1]

traceArr = np.zeros([len(pauliMatrices), len(pauliMatrices)])

for i in range(len(pauliMatrices)):
    for j in range(len(pauliMatrices)):
        traceArr[i,j] = np.real(np.trace(np.mat(pauliMatrices[i]) * np.mat(pauliMatrices[j])))
        
plt.pcolor(traceArr)
plt.colorbar()
plt.show()
```
