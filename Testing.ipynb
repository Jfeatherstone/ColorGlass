```python
from TwoColors import Proton, Nucleus
import matplotlib.pyplot as plt
import numpy as np
from Collision import *
```

```python
N = 128
delta = .1
mu = 2
radius = 1

nucleus = Nucleus(N, delta, mu)
proton = Proton(N, delta, mu, radius)
proton2 = Proton(N, delta, mu, radius)

c1 = Collision(nucleus, proton)

plt.plot(c1.momentaBins(), c1.particlesProduced())
plt.yscale('log')
```

```python
N = 128
delta = .1
mu = 2
radius = 1

nucleus = Nucleus(N, delta, mu) 
proton = Proton(N, delta, mu, radius)
col = Collision(nucleus, proton)

nucleus.colorChargeField()
nucleus.gaugeField()
nucleus.wilsonLine()
nucleus.adjointWilsonLine()

proton.colorChargeField()
proton.gaugeField()

col.omega()
col.omegaFFT()
col.momentaBins()
col.particlesProduced()

plt.plot(col.momentaBins(), col.particlesProduced())
```

```python
fig, ax = plt.subplots(2, gluonDOF, figsize=(12, 6))
for i in range(nucleus.gluonDOF):
    ax[0,i].pcolor(nucleus.colorChargeField()[i])
    ax[1,i].pcolor(nucleus.gaugeField()[i])
fig.tight_layout()
plt.show()

fig, ax = plt.subplots(2, gluonDOF, figsize=(12, 6))
for i in range(proton.gluonDOF):
    ax[0,i].pcolor(proton.colorChargeField()[i])
    ax[1,i].pcolor(proton.gaugeField()[i])
fig.tight_layout()
plt.show()
```
